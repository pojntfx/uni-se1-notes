# Merry Christmas

```java
class TreePrinter {
    private String getCharLine(int width, String toAdd) {
        var out = "";
        
        for (var i = 0; i < width; i++) {
            out += toAdd;
        }
        
        return out;
    }
    
    private String getCenteredLine(int totalWidth, int lineWidth, String toAdd) {
        var whitespaceWidth = (totalWidth - lineWidth)/2;
        var whitespace = getCharLine(whitespaceWidth, " ");
        
        return String.format("%s%s%s", whitespace, getCharLine(lineWidth, toAdd), whitespace);
    }
    
    public String getTree(int width, int height, int trunkHeight) {
        var star = getCenteredLine(width, 1, "X");
        
        var body = "";
        for (var i = 1; i <= width; i++) {
            if (i % 2 != 0) {
                body += String.format("%s%s", i == width ? "" : "\n", treePrinter.getCenteredLine(width, i, "*"));
            }
        }
        
        var footer = "";
        for (var i = 0; i < trunkHeight; i++) {
            footer += String.format("%s%s", i == trunkHeight ? "" : "\n", treePrinter.getCenteredLine(width, 3, "#"));
        }
        
        return String.format("%s%s%s", star, body, footer);
    }
}
```

```java
var treePrinter = new TreePrinter()
```

```java
treePrinter.getTree(10, 5, 2)
```
